```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('Placement.csv')
df.head()
```

```python
X = df.iloc[:,1].values
y = df.iloc[:,2].values
```

```python
print(X)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, random_state = 2, test_size=0.2)
```

```python
class LR:
  def __init__(self):
    self.m = None
    self.b = None

  def _fit(self, X_train, y_train):

    x_mean = X_train.mean()
    y_mean = y_train.mean()

    num = den = 0

    for i in range(X_train.shape[0]):

      num += (y_train[i] - y_mean) * (X_train[i] - x_mean)
      den += (X_train[i] - x_mean) * (X_train[i] - x_mean)

    self.m = num / den
    self.b = y_mean - (x_mean * self.m)

    return [self.m, self.b]

  def _predict(self, test):

    return (self.m * test) + self.b

```

```python
lr = LR()
```

```python
m,b = lr._fit(X_train, y_train)
```

```python
print(m,b)
```

```python
lr._predict(8.14)
```

```python
y_input = m*X + b
```
